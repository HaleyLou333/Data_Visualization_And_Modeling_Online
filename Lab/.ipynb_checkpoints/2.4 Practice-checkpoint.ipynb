```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("titanic_passengers.csv")

# Display the first few rows to confirm it loaded correctly
df.head()
```

```python
df.groupby("Pclass")["Age"].agg(["mean", "median"])
```

```python
plt.figure(figsize=(10,6))

sns.histplot(data=df[df["Pclass"] == 1], x="Age", color="blue", label="1st Class", kde=False, alpha=0.5)
sns.histplot(data=df[df["Pclass"] == 2], x="Age", color="green", label="2nd Class", kde=False, alpha=0.5)
sns.histplot(data=df[df["Pclass"] == 3], x="Age", color="red", label="3rd Class", kde=False, alpha=0.5)

plt.title("Overlaid Age Distributions by Passenger Class")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend()
plt.show()
```

Create a categorical version of Pclass

```python
df['Pclass_cat'] = df['Pclass'].astype('str')
```

Plot the boxplots

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Pclass_cat', y='Age')

plt.title("Age Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.show()
```

### What is one feature of the difference in the distribution of age between passenger ticket classes that you find easier to observe from the overlaid histogram?  What is one feature that's easiest to see from the side-by-side boxplots.

From the overlaid histogram, it's easiest to see overall shape of the age distributions for each passenger class. For example, that 1st class passengers tend to cluster more in the middle-aged range while 3rd class passengers have more children and young adults.

For the side-by-side boxplots, it's easiest to compare the medians, quartiles, and outliers across classes. The boxplots make it clear that 1st class passengers have the highest median age and 3rd class the lowest, and they show differences in the IQR that are harder to see in the histogram.

### Describe the differences in the center, shape, and spread of age for the different passenger classes.

The center of the age distribution differs across classes: 1st class passengers have the highest median age, 2nd class passengers fall in the middle, and 3rd class passengers have the lowest median age. The shapes of the distributions also differ: 1st class is more concentrated around middle-aged adults, while 3rd class is more right-skewed with many children and young adults. The spread is widest in 3rd class, which includes a broad range of ages from infants to older adults, whereas 1st class shows a more moderate spread.
