```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Sleep.csv")
df.head()
```

### Scatterplot: Non-dreaming Sleep vs Total Sleep

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="SWsleep", y="Totsleep")
plt.title("Non-dreaming Sleep vs Total Sleep")
plt.xlabel("SWsleep (hrs)")
plt.ylabel("Total Sleep (hrs)")
plt.show()
```

### Interpretation: SWsleep vs Total Sleep
Mammals that sleep more in non-dreaming stages generally sleep more overall. The relationship is postive and approximately liner, though not perfectly tight. 

### Add Color Coding for Predation Index

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="SWsleep", y="Totsleep", hue="Pred", palette="viridis")
plt.title("Non-dreaming Sleep vs Total Sleep by Predation Index")
plt.xlabel("SWsleep (hrs)")
plt.ylabel("Total Sleep (hrs)")
plt.show()
```

### Interpretation:
Species that are more vulnerable to predators tend to sleep fewer hours, while low-risk species sleep longer. Sleep duration appears to depend partly on predation risk.

```python
### Body Weight vs Total Sleep
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Body", y="Totsleep")
plt.xscale("log")  # optional but helpful due to huge weight range
plt.title("Body Weight vs Total Sleep")
plt.xlabel("Body Weight (kg, log scale)")
plt.ylabel("Total Sleep (hrs)")
plt.show()
```

### Interpretation:
A liner model is not a good fit. The relationship is clearly nonliner, with small mammals sleeping for more than large mammals. A curved or log-transformed model would better capture the pattern.
