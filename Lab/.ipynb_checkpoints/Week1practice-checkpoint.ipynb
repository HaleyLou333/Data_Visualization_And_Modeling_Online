```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look a little nicer
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("titanic_passengers.csv")
df.head()
```

```python
df.info()
```

```python
df.head(10)
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['Fare'], bins=30, edgecolor='black')
plt.title('Distribution of Titanic Passenger Fares')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
# Most passengers paid relatively low fares, and a small number paid very high fares.
df['Fare'].median()
```

```python
# Typical ticket prices were low at 14.45
```

```python
df['Fare'].min(), df['Fare'].max()
```

```python
# The distribution of passenger fares is strongly right-skewed. Most passengers paid relatively low fares, typically under $50, with a median around $14.
# The fare range spans from $0 to about %512, indicating that while most passengers traveled cheaply, a small number of first-class passengers paid extremely high prices for their accomodations.
```

```python
survival_counts = df['Survived'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(survival_counts.index, survival_counts.values, color=['red', 'green'])
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.ylabel('Number of Passengers')
plt.title('Survival Counts on the Titanic')
plt.show()
```

```python
survival_rate = df['Survived'].mean()
survival_rate
```

```python
sex_counts = df['Sex'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    sex_counts,
    labels=sex_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['lightblue', 'lightpink']
)
plt.title('Proportion of Male and Female Passengers')
plt.show()
```
