## Open new notebook and load the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic_passengers.csv")
df.head(10)
```

## Create a subset of only 3rd-class passengers

```python
third_class = df[df['Pclass'] == 3]
third_class.head(10)
```

## Using .groupby(), calculate the mean age of survivors vs. non-survivors

```python
df.groupby('Survived')['Age'].mean()
```

## Create a new feature: Fare in Canadian Dollars

### Conversion: 1 British pound = 1.70 Canadian dollars

```python
df['Fare in CAD'] = df['Fare'] * 1.70
df[['Fare', 'Fare in CAD']].head()
```

## Create a crosstab of 'Pclass' and 'Survived'

```python
pd.crosstab(df['Pclass'], df['Survived'])
```

Interpretation: Passengers in 3rd class were least likely to survive, which matches history.

## Create a new feature called 'Port_name'

```python
df['Port_name'] = df['Embarked'].map({
    'S': 'Southampton',
    'C': 'Cherbourg',
    'Q': 'Queenstown'
})
```

```python
df[['Embarked', 'Port_name']].head()
```
