# Lab: Spatial and Time-Series Visualizations Using Gapminder
This lab creates three spatial + time-series visualizations using the Gapminder dataset. The dataset includes country-level information on life expectancy, GDP per capita, and population from 1952-2007.

```python
import plotly.express as px
import pandas as pd

df = px.data.gapminder()
df.head
```

## Dataset Description
The Gapminder dataset comes from the Gapminder Foundation and is included directly in Plotly Express. It contains observations for countries across multiple decades, allowing analysis of how development indicators change over both space and time.

**Features included**
- 'country': Country name
- 'continent': Contintent grouping
- 'year': Observation year
- 'lifeExp': Life expectancy at birth
- 'pop': Population
- 'gdpPercap': GDP per capita
- 'iso_alpha'/'iso_num': Country codes used for mapping

This dataset is ideal for spatial and temporal visualization because it tracks global development across many years and regions.

## Visualization 1: Life Expectancy Over Time
This choropleth shows how life expectancy changes across countries from 1952-2007.

```python
fig = px.choropleth(
    df,
     locations="iso_alpha",
    color="lifeExp",
    hover_name="country",
    animation_frame="year",
    color_continuous_scale="Viridis",
    projection="natural earth"
)
fig.show()
```

Life expectancy increases steadily across most reigons over time, especially in Asia and South America. Spatially, Africa remains lower for longer, showing uneven global improvements in health.

## Visualization 2: GDP per Capita Over Time
This map shows how economic development changes across countries over the same time period.

```python
fig = px.choropleth(
    df,
      locations="iso_alpha",
    color="gdpPercap",
    hover_name="country",
    animation_frame="year",
    color_continuous_scale="Viridis",
    projection="natural earth"
)
fig.show()
```

GDP per capita rises sharply in Europe, North America, and parts of Asia. Many African countries remain in lower economic brackets, highlighting persistent global inequality.

## Visualization 3: Population Over Time
This choropleth visualizes how population changes across countries from 1952-2007.

```python
df_year_sorted = df.sort_values(by="year")
fig = px.choropleth(
    df_year_sorted,
    locations="iso_alpha",
    color="pop",
    hover_name="country",
    animation_frame="year",
    color_continuous_scale="Viridis",
    projection="natural earth"
)
fig.show()
```

Populaton grows dramatically in Asia, especially China and India, which dominate the global population landscape. Spatially, population remains concentratedin a few regions, while many countries stay lightly populated.

## Summary
These three visualizations highlight how life expectancy, GDP per capita, and population evolve across both time and geography. Together, they illustrate global development patterns and regional disparities over the second half of the 20th century, but they also reveal persistent inequalities: wealth and health improvements occur much faster in Europe, North America, and parts of Asia, while many African and lower-income contries experience slower progress. By comparing these indicators across space and time, the visualizations make it easy to see both global advancement and the uneven distribution of those gains.
