```python
import plotly.express as px
import pandas as pd

df = px.data.gapminder()
df.head()
```

```python
df.columns
```

### Recreate the lecture choropleth (template)

```python
fig = px.choropleth(
    df,
    locations="iso_alpha",
    color="lifeExp",
    hover_name="country",
    animation_frame="year",
    color_continuous_scale="Viridis",
    projection="natural earth"
)
fig.show()
```

### Change the color to GDP per capita

```python
fig = px.choropleth(
    df,
    locations="iso_alpha",
    color="gdpPercap",
    hover_name="country",
    animation_frame="year",
    color_continuous_scale="Viridis",
    projection="natural earth"
)
fig.show()
```

### Color = life expectancy, Animation = GDP per capita, AND sort the data by GDP first

```python
df_gdp_sorted = df.sort_values(by="gdpPercap")

fig = px.choropleth(
    df_gdp_sorted,
    locations="iso_alpha",
    color="lifeExp",
    hover_name="country",
    animation_frame="gdpPercap",
    color_continuous_scale="Viridis",
    projection="natural earth"
)
fig.show()
```

### Animation frame back to year, Color -> population (pop), Sort by year first

```python
df_year_sorted = df.sort_values(by="year")

fig = px.choropleth(
    df_year_sorted,
    locations="iso_alpha",
    color="pop",
    hover_name="country",
    animation_frame="year",
    color_continuous_scale="Viridis",
    projection="natural earth"
)
fig.show()
```

### Conclusion:
The life expectancy map animated by year provides the clearest and most meaningful vizualizations of global development. It shows long-term improvements in health while still reflecting underlying economic patterns, making it most effective for understanding worldwide trends in GDP, life expectancy, and population.
