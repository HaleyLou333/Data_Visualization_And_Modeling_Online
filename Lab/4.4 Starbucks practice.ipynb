```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Starbucks_reviews.csv")
df.head()
df.info()
```

## Visualize missing data

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False)
plt.title("Missing Data in Starbucks Reviews")
plt.show()
```

```python
df.isnull().sum()
```

## Visualize the "Rating" feature

```python
plt.figure(figsize=(7,4))
sns.countplot(x="Rating", data=df, palette="viridis")
plt.title("Distribution of Starbucks Review Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
df["Rating"].describe()
```

## Make at least one text visualization

```python
from wordcloud import WordCloud, STOPWORDS

text = " ".join(df["Review"].dropna())

wc = WordCloud(
    width=800,
    height=400,
    background_color="white",
    stopwords=STOPWORDS
).generate(text)

plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("Most Common Words in Starbucks Reviews")
plt.show()
```

Starbucks reviews tend to be emotionally polarized: very happy customers or very frustrated ones, with fewer neutral reviews.
