### Dataset: Palmer Penguins 
The Palmer Penguins dataset contains measurements for 344 penguins collected by researchers at Palmer Station, Antartica.

### Target Feature
**body_mass_g** - the penguin's body mass in grams

### Predictor Features
1. **bill_length_mm** - bill length in millimeters
2. **bill_depth_mm** - bill depth in millimeters
3. **flipper_length_mm** - flipper length in millimeters
4. **species** - Adelie, Chinstrap, or Gentoo
5. **sex** - male or female

These features were selected because they represent biologically meaningful measurements that may influence or relate to body mass.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")
penguins.head()
```

### Bill Length vs Body Mass

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=penguins, x="bill_length_mm", y="body_mass_g", hue="species")
plt.title("Bill Length vs Body Mass")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Body Mass (g)")
plt.show()
```

Bill length shows a postivie relationship with body mass. Penguins with longer bills tend to weigh more overall. The trend is especially clear within species groups, suggesting that bill length is a meaningful size indicator.

### Bill Depth vs Body Mass

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=penguins, x="bill_depth_mm", y="body_mass_g", hue="species")
plt.title("Bill Depth vs Body Mass")
plt.xlabel("Bill Depth (mm)")
plt.ylabel("Body Mass (g)")
plt.show()
```

Bill depth does not show a strong linear relationship with body mass. Instead, the pattern varies by species: Adelie penguins have deeper bills but are lighter than Gentoo penguins. This suggests bill depth is more species-specific than mass-related.

### Flipper Length vs Body Mass

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
plt.title("Flipper Length vs Body Mass")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
plt.show()
```

Flipper length shows the strongest postive relationship with body mass. Heavier penguins consistently have longer flippers, and the trend is clear across all species. This makes flipper length one of the best predictors of body mass in the dataset.

### Species vs Body Mass

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=penguins, x="species", y="body_mass_g")
plt.title("Body Mass by Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()
```

Species has a strong influence on body mass. Gentoo penguins are the heaviest on average, Chinstrap are in the middle, and Adelie are the lightest. The differences between species are large and consistent.

### Sex vs Body Mass

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=penguins, x="sex", y="body_mass_g")
plt.title("Body Mass by Sex")
plt.xlabel("Sex")
plt.ylabel("Body Mass (g)")
plt.show()
```

Sex shows a moderate relationship with body mass. Male penguins tend to weigh more than females across species, although the difference is smaller than the species effect.

### Big Picture Conclusion
Across all five visualizations, several clear patters emerged about what influences penguins body mass. Flipper length showed the strongest positive relationship with body mass, indicating that larger penguins consistently have longer flippers. Bill length also increased with body mass, although the relationship was slightly weaker. Bill depth did not show a consistent trend with mass and appeared to vary more by species than by size. Species differences were substantial: Gentoo penguins were the heaviest, Chinstrap were intermediate, and Adelie were the lightest. Sex also played a role, with males generally weighing more than females across species. Overall, the data suggest that body mass is shaped by both physical size measurements and species-level differences. Flipper length, species, and bill length are the most informative predictors of body mass in the Palmer Penguins dataset.
